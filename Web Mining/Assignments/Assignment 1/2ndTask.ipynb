```python
import nltk  
import numpy as np  
import random  
import string

import bs4 as bs  
import urllib.request  
import re  
```

```python
wordfreq1 = {}
wordfreq2 = {}
for sentence in readwords1:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq1.keys():
            wordfreq1[token] = 1
        else:
            wordfreq1[token] += 1
for sentence in readwords2:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq2.keys():
            wordfreq2[token] = 1
        else:
            wordfreq2[token] += 1
```

```python
wordfreq1,wordfreq2
```

```python
import heapq
most_freq1 = heapq.nlargest(200, wordfreq1, key=wordfreq1.get)
most_freq2 = heapq.nlargest(200, wordfreq2, key=wordfreq2.get)
```

```python
sentence_vectors1 = []
sentence_vectors2 = []
for sentence in readwords1:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq1:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors1.append(sent_vec)
for sentence in readwords2:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq2:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors2.append(sent_vec)
```
